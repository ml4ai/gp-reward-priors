```python
import pandas as pd

pd.options.mode.copy_on_write = True
```

```python
antmaze_64_2 = pd.read_csv(
    "../data/antmaze/tuning_results/antmaze_large_diverse_64_2.csv"
)

for i in antmaze_64_2:
    if "MIN" in i or "MAX" in i:
        antmaze_64_2 = antmaze_64_2.drop(i, axis=1)

antmaze_64_2 = antmaze_64_2.drop("Step", axis=1).abs()
antmaze_64_2_min_idx = antmaze_64_2.min().idxmin()
print(antmaze_64_2_min_idx)
print((antmaze_64_2.idxmin() + 1)[antmaze_64_2_min_idx])
```
